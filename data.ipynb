```python
import pandas as pd
import pandas as pd
import sqlalchemy
from sqlalchemy import types
```

```python
# Connect to MySQL
engine = sqlalchemy.create_engine("mysql+pymysql://root:anukool@localhost:3306/godrej_estac")
```

```python
df = pd.read_csv('./estac_records.csv')
```

```python
df['device_time'] = pd.to_datetime(df['device_time'])
```

```python
# Upload DataFrame to SQL
df.to_sql(
    name="estac_data",
    con=engine,
    if_exists="append",
    index=False,
    dtype={
        "device_time": types.DateTime(),
        "humidity": types.Numeric(5,2),
        "temperature": types.Numeric(5,2)
    }
)

print("Data inserted successfully!")
```
